```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_edge("Sarah", "Emily", weight = 1)
G.add_edge("Sarah", "Christopher", weight = 2)
G.add_edge("Sarah", "Andrew", weight = 3)

G.add_edge("Stuart", "Jacob", weight = 4)

G.add_edge("Jessica", "Ashley", weight = 5)

G.add_edge("Jacob", "Christopher", weight = 6)

G.add_edge("Christopher", "Andrew", weight = 7)
G.add_edge("Christopher", "Ashley", weight = 8)

G.add_edge("Ashley", "Emily", weight = 9)
G.add_edge("Ashley", "Joshua", weight = 10)

G.add_edge("Joshua", "Emily", weight = 11)
G.add_edge("Sarah", "Andrew", weight = 12)

pos = {"Stuart": (3, 4), "Jessica": (4, 4), "Sarah": (1, 3), "Jacob": (3, 3), "Ashley": (4, 3), "Andrew": (2, 1), "Christopher":(2.5, 2), 
       "Emily":(4, 2), "Joshua":(4.5, 1)}

nx.draw(G, pos ,with_labels = True, node_color = "lightblue", edge_color = "black")
edge_labels = {(u, v): f"{d['weight']}" for u, v, d in G.edges(data = True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels = edge_labels)

plt.show()

print(G.number_of_nodes())
print(G.number_of_edges())
print(G.nodes())
print(G.edges())
```
