```python
"""
Task 2: Friendship Network
Date: February 24, 2025
"""

import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

edges = [
    ("Ashley", "Christopher"), ("Ashley", "Emily"), ("Ashley", "Joshua"),
    ("Bart", "Lisa"), ("Bart", "Matthew"), ("Christopher", "Andrew"),
    ("Emily", "Joshua"), ("Jacob", "Christopher"), ("Jessica", "Ashley"),
    ("JorEl", "Zod"), ("KalEl", "JorEl"), ("Kyle", "Lex"), ("Kyle", "Zod"),
    ("Lisa", "Marge"), ("Matthew", "Lisa"), ("Michael", "Christopher"),
    ("Michael", "Joshua"), ("Michael", "Jessica"), ("Samantha", "Matthew"),
    ("Samantha", "Tyler"), ("Sarah", "Andrew"), ("Sarah", "Christopher"),
    ("Sarah", "Emily"), ("Tyler", "Kyle"), ("Stuart", "Jacob")
        ]

G.add_edges_from(edges)

pos = {
"Ashley": (0, 4),
"Christopher": (1, 3),
"Emily": (0, 2),
"Joshua": (2, 2),
"Bart": (-4, 4),
"Lisa": (-2, 3),
"Matthew": (-2, 5),
"Andrew": (4, 4),
"Jacob": (3, 6),
"Jessica": (1, 6),
"JorEl": (6, 2),
"Zod": (7, 1),
"KalEl": (6, 3),
"Kyle": (4, 0),
"Lex": (6, -1),
"Marge": (-3, 2),
"Michael": (2, 6),
"Samantha": (-1, 6),
"Tyler": (-1, 8),
"Sarah": (2, 1),
"Stuart": (4, 7),
}
```

```python
# Compute shortest path distances for the example pairs
def print_friendship_distances(G, pairs):
    for u, v in pairs:
        try:
            distance = nx.shortest_path_length(G, u, v)
            path = nx.shortest_path(G, u, v)
            print(f"Distance({u}, {v}) = {distance}, Path: {' -> '.join(path)}")
        except nx.NetworkXNoPath:
            print(f"No path between {u} and {v}")

names = ["Ashley", "Christopher", "Emily", "Joshua", "Bart", "Lisa", "Matthew", "Andrew", "Jacob", "Jessica", 
         "JorEl", "Zod", "KalEl", "Kyle", "Lex", "Marge", "Michael", "Samantha", "Tyler", "Sarah", "Stuart"]

# Test pairs as an alternative way of testing the print_friendship_distances() function
test_pairs = [("Ashley", "Christopher"),
              ("Ashley", "Andrew"),
              ("Ashley", "Stuart")]

# User input that allows testing of any name in our list
print("This is a program that calculates friendship distances. Choose a pair of names out of this list:\n", names)
user_input1 = input("Name 1: ")
user_input2 = input("Name 2: ")
user_pair = [(user_input1, user_input2)]
print_friendship_distances(G, user_pair)
```

```python
nx.draw(G, pos, with_labels=True, node_color="lightgreen", edge_color="black",
        node_size=700, font_size=8)
plt.title("Friendship Network")

plt.show()

# Print graph stats
print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
print(f"Nodes: {list(G.nodes())}")
print(f"Edges: {list(G.edges())}")
```
